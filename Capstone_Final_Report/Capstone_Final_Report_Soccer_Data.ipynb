**3.2.3 Difference in Ratings Between Positions**

To predict the player rating we can separate  the data into train and test datasets and fit a regression model to the data to find a model with a good fit. However, we know that the data has 36 attributes for each player and according to our assumptions about the game, we think that there is a difference between ratings based on which position the player plays at. If this is true, it would make sense for us to separate  the data by positions and fit separate  models on each position data. For this, we test the following hypothesis:

H0: There is no significant difference between the mean ratings of the different position groups
                                        VS
H1: There is significant difference between the mean ratings of the different position groups

```python
#subset the data for all positions
data = players_agg[['player_position', 'overall_rating']]
defn = data[data.player_position == 'Defender']
forwd = data[data.player_position == 'Forward']
midfld = data[data.player_position == 'Midfielder']
gkeep = data[data.player_position == 'Goalkeeper']
#Create a boxplot
data.boxplot('overall_rating', by='player_position', figsize=(12, 8))
stats.f_oneway(defn.overall_rating, forwd.overall_rating, midfld.overall_rating, gkeep.overall_rating)
```

The test result suggests the groups don't have the same sample means in this case, since the p-value is significant at a 95% confidence level. A common post hoc-test is Tukey's test, we carry out Tukey's test using the pairwise_tukeyhsd() function in the statsmodels.stats.multicomp library:

```python
# post hoc analysis to identify difference between groups
tukey = pairwise_tukeyhsd(endog=data.overall_rating,     # Data
                          groups=data.player_position,   # Groups
                          alpha=0.05)          # Significance level

tukey.plot_simultaneous()    # Plot group confidence intervals
plt.vlines(x=49.57,ymin=-0.5,ymax=4.5, color="red")

tukey.summary()      
```

The output of the Tukey test shows the average difference, a confidence interval as well as whether you should reject the null hypothesis for each pair of groups at the given significance level. In this case, the test suggests we reject the null hypothesis for 5 pairs, with an exception of the groups Defender-Goalkeeper. The 95% confidence interval plot reinforces the results visually: only Defender and Goalkeeper groups' confidence intervals overlap.

We have enough evidence to conclude that there is difference in ratings between the positional groups and we go forward with our suggestion to fit separate models for each position.

*Define function to identify most and least correlating features for given data*

```python
#function to measure relationship between target variable and each of the features independently
def feature_correlation(position_df):
    #function output is dataframe of r,r2,p-value of target variable with features
    #get correlation of features with overall rating in dataframe
    corr_df = position_df.corr()
    corr_df = corr_df.iloc[:, 1]
    corr_df = pd.DataFrame(corr_df.iloc[2:, ])
    corr_df.reset_index(level=0, inplace=True)
    corr_df.rename(columns={'index': 'features', 'overall_rating': 'r_score'}, inplace=True)
    #get r, rsquare and pvalue of overall rating vs features to find significant features
    features_list =['age', 'height',  'weight', 'finishing',  'heading_accuracy',  'short_passing',  'volleys',  'dribbling',  'curve',  'free_kick_accuracy',  'long_passing',  'ball_control',  'acceleration',  'sprint_speed',  'agility',  'reactions',  'balance',  'shot_power',  'jumping',  'stamina',  'strength',  'long_shots',  'aggression',  'interceptions',  'positioning',  'vision',  'penalties',  'marking',  'standing_tackle',  'sliding_tackle', 'gk_diving',  'gk_handling',  'gk_kicking', 'gk_positioning', 'gk_reflexes']
    target = position_df.ix[:,3:4].values
    score = []
    #len_count = len(count)
    for i in range(4, 39):
        lm.fit(position_df.ix[:,i:i+1].values, target)
        val = lm.score(position_df.ix[:,i:i+1].values, target)
        score.append(val)

    rsquared_df = pd.DataFrame({'features': features_list,'r2_score': score})
    #get significance p value:
    def calculate_pvalues(df):
        #function to calculate significance p value
        df = df.dropna()._get_numeric_data()
        dfcols = pd.DataFrame(columns=df.columns)
        pvalues = dfcols.transpose().join(dfcols, how='outer')
        for r in df.columns:
            for c in df.columns:
                pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
        return pvalues
    p_values_df = calculate_pvalues(position_df)
    p_values  = p_values_df.iloc[:, 1]
    p_values = pd.DataFrame(p_values.iloc[2:, ])
    p_values.reset_index(level=0, inplace=True)
    p_values.rename(columns={'index': 'features', 'overall_rating': 'pvalue'}, inplace=True)
    #merge all three dataframes
    feature_significance_df = corr_df.merge(rsquared_df, on='features')
    feature_significance_df = feature_significance_df.merge(p_values, on='features')
    feature_significance_df = feature_significance_df.sort_values(['r_score'], ascending = False)
    return(feature_significance_df)
```

*Define Functions to fit Regression Models*

```python
#Defining the functions:
model_scores = {}

#scree plot function to plot pca components explaining variance in data:
def scree_plot_pca(n_components, data):
    #the function plots a scree plot of pca components of given data which allows us to select an optimal no of components
    pca = PCA(n_components)
    pca.fit(data)
    var= pca.explained_variance_ratio_
    var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
    plt.plot(var1)
    show = plt.show()
    return(show)


#linear regression scores
def linear_regression_fit(x_train, y_train, x_test, y_test):
    #fit a linear regression on the data and get accuracy score of linear regression model
    lm.fit(x_train, y_train)
    y_pred = lm.predict(x_test)
    train_score = lm.score(x_train, y_train)
    test_score = lm.score(x_test, y_test)
    rss = sum((y_pred-y_test)**2)
    rms = sqrt(mean_squared_error(y_test, y_pred))
    print("LR Train Accuracy Score = {:.2f}".format(train_score*100))
    print("LR Test Accuracy Score = {:.2f}".format(test_score*100))
    print("LR Residual Sum of Squares = %d" % rss)
    print("LR Root Mean Squre Error = {:.2f}" .format(rms))
    model_scores['Linear_Regression'] = format(test_score*100)
    return()


#polynomial model fit regression scores
def polynomial_regression_fit(degree, x_train, y_train, x_test, y_test):
    #fit a polnomial regression on the data and get accuracy score of polynomial regression model
    poly = PolynomialFeatures(degree)
    vector = y_train
    predict = x_test
    #transform the x data for proper fitting (for single variable type it returns,[1,x,x**2]
    X_ = poly.fit_transform(x_train)
    #transform the prediction to fit the model type
    predict_ = poly.fit_transform(predict)
    #here we can remove polynomial orders we don't want
    #for instance I'm removing the `x` component
    X_ = np.delete(X_,(1),axis=1)
    predict_ = np.delete(predict_,(1),axis=1)
    #preform the actual regression
    lm.fit(X_, vector)
    y_pred = lm.predict(predict_)
    train_score = lm.score(X_, vector)
    test_score = lm.score(predict_, y_pred)
    rss = sum((y_pred-y_test)**2)
    rms = sqrt(mean_squared_error(y_test, y_pred))
    print("Polynomial Regression Train Accuracy Score = {:.2f}".format(train_score*100))
    print("Polynomial Regression Test Accuracy Score = {:.2f}".format(test_score*100))
    print("Polynomial Regression Residual Sum of Squares = %d" % rss)
    print("Polynomial Regression Root Mean Squre Error = {:.2f}" .format(rms))
    return()


def ridge_regression_fit(alpha, x_train, y_train, x_test, y_test):
    #fit a ridge regression on the data and get accuracy score of ridge regression model
    clf = Ridge(alpha)
    clf.fit(x_train, y_train)
    train_score = clf.score(x_train, y_train)
    test_score = clf.score(x_test, y_test)
    y_pred = clf.predict(x_test)
    rss = sum((y_pred-y_test)**2)
    rms = sqrt(mean_squared_error(y_test, y_pred))
    print("Ridge Train Accuracy Score = {:.2f}".format(train_score*100))
    print("Ridge Test Accuracy Score = {:.2f}".format(test_score*100))
    print("Ridge Residual Sum of Squares = %d" % rss)
    print("Ridge Root Mean Squre Error = {:.2f}" .format(rms))
    model_scores['Ridge_Regression'] = format(test_score*100)
    return()


def lasso_regression_fit(alpha, x_train, y_train, x_test, y_test):
    #fit a lasso regression on the data and get accuracy score of lasso regression model
    clf = Lasso(alpha)
    clf.fit(x_train, y_train)
    train_score = clf.score(x_train, y_train)
    test_score = clf.score(x_test, y_test)
    
    pred_LSR = clf.predict(x_test)
    a = []
    a.append(pred_LSR)
    a = np.array(a)
    a.flatten()
    
    
    rss = sum((a.T-y_test)**2)
    rms = sqrt(mean_squared_error(y_test, pred_LSR))
    print("Lasso Train Accuracy Score = {:.2f}".format(train_score*100))
    print("Lasso Test Accuracy Score = {:.2f}".format(test_score*100))
    print("Lasso Residual Sum of Squares = %d" % rss)
    print("Lasso Root Mean Squre Error = {:.2f}" .format(rms))
    model_scores['Lasso_Regression'] = format(test_score*100)
    return()


def svr_fit(x_train, y_train, x_test, y_test):
    #fit a svr model on the data and get accuracy score of svr model
    svr = SVR(C=10.0, epsilon=0.2, gamma=0.01, kernel='rbf')
    svr.fit(x_train,y_train)
    
    pred_SVR = svr.predict(x_test)
    a = []
    a.append(pred_SVR)
    a = np.array(a)
    a.flatten()
    
    svr_train_score = svr.score(x_train, y_train)
    svr_test_score = svr.score(x_test, y_test)
    
    rss_svr = sum((a.T-y_test)**2)
    rms_svr = sqrt(mean_squared_error(y_test, pred_SVR))
    
    print("SVR Train Accuracy Score = {:.2f}".format(svr_train_score*100))
    print("SVR Test Accuracy Score = {:.2f}".format(svr_test_score*100))
    print("SVR Residual Sum of Squares = %d" % rss_svr)
    print("SVR Root Mean Squre Error = {:.2f}" .format(rms_svr))
    model_scores['SVR'] = format(svr_test_score*100)
    return()

''' #param selection for svm took very long time to run, so excluded that from code after selecting best params
def svr_param_selection(X, y):
    #function to select best parameters for svr model, note takes a long time to run
    kernels = ['linear', 'rbf','poly']
    Cs = [0.001, 0.01, 0.1, 1, 10, 100]
    gammas = [0.001, 0.01, 0.1, 1]
    epsilon = [0.1,0.2,0.5,0.3]
    param_grid = {'kernel': kernels, 'C': Cs, 'gamma' : gammas, 'epsilon' : epsilon}
    grid_search = GridSearchCV(svm.SVR(), param_grid)
    grid_search.fit(X, y)
    grid_search.best_params_
    return(grid_search.best_params_)
'''
# define function for calculating cross validation scores for a model
def kfold_cross_val_score(clf, x_train, y_train):
    start_time = timeit.default_timer()
    #function to calculate accuracy on kfold cross validation for a given model, define model as clf prior to calling function
    kfold = KFold(n_splits=10, shuffle=True, random_state=0)
    scores = cross_val_score(clf, x_train, y_train, cv=kfold)
    predicted = cross_val_predict(clf, x_train, y_train, cv=kfold)
    elapsed = timeit.default_timer() - start_time
    print('Time elapsed: = {:.2f} '.format(elapsed))
    print("Cross-validation scores:\n{}".format(scores))
    print("Accuracy: %0.4f (+/- %0.4f)" % (scores.mean(), scores.std() * 2))
    return()

def model_scores_table(model_scores):
    model_scores_x = pd.DataFrame(list(model_scores.items()), columns=['Model', 'Test_Accuracy'])
    model_scores_x.Test_Accuracy = round(model_scores_x.Test_Accuracy.astype(float),2)
    return(model_scores_x)
```

## ** 9. Limitations **

There are certain limitations in the data that reduce the robustness of the models we have developed:

1. Not all historical data is available for each player. Knowing that the FIFA(game) Player Ranking is based on historical data of player, we do not have this data for each player. A lot of information will not be learned by the model due to the absence of this data.
2. Data related to ranks of leagues, teams and countries is also not available in this data. This information would have made the model more accuracte as these factors play an important role in determining player ranking.
3. There is not enough historical data available for each player, if it were, a time series model could be built to predict the future ratings of players
4. There was country and league data missing for few players and also more than 20% players from our players data had not been active in the last two years. These players were excluded from featuring in the top players lists.

**Additional: Find top 10 players for each position below 25 years of age**

```python
def top_x_players_under_25(data, x):
    data = data[(data.age < 25)]
    df = data.sort_values(['predicted_overall_rating'], ascending=False).head(x)
    df['player_rank'] = df.predicted_overall_rating.rank(ascending=False)
    df = df[['player_api_id', 'player_position', 'player_name', 'predicted_overall_rating', 'overall_rating', 'player_rank', 'league_name', 'country_name', 'age']]     
    df.player_rank = df.player_rank.astype(int)
    df.overall_rating = df.overall_rating.astype(int)
    df.age = df.age.astype(int)
    return(df)
```
